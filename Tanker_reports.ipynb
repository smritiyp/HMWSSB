```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = '/content/merged_reports.csv'
data = pd.read_csv(file_path)

# Assuming 'delivered' is the column you want to analyze for anomalies
data['delivered'] = pd.to_numeric(data['delivered'], errors='coerce')
data.dropna(subset=['delivered'], inplace=True)

# Calculate mean and standard deviation
mean_delivered = data['delivered'].mean()
std_delivered = data['delivered'].std()

# Define a threshold for anomalies (e.g., 3 standard deviations from the mean)
threshold = 3

# Identify anomalies
anomalies = data[(data['delivered'] > mean_delivered + threshold * std_delivered) |
                 (data['delivered'] < mean_delivered - threshold * std_delivered)]

# Plot the data and highlight anomalies
plt.figure(figsize=(10, 6))
plt.plot(data['year'], data['delivered'], label='Delivered Tankers')
plt.scatter(anomalies['year'], anomalies['delivered'], color='red', label='Anomalies')
plt.xlabel('Year')
plt.ylabel('No of Tankers delivered')
plt.title('Tankers report (with Anomalies)')
plt.xticks(rotation=45)
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
df = pd.read_csv('/content/merged_reports.csv')
df['year'] = pd.to_numeric(df['year'], errors='coerce')
df['delivered'] = pd.to_numeric(df['delivered'], errors='coerce')
df = df.dropna(subset=['year', 'delivered'])
jitter = np.random.normal(0, 0.1, len(df))
x_jittered = df['year'] + jitter
plt.scatter(x_jittered, df['delivered'], alpha=0.5)
plt.xticks([2022, 2023, 2024])
plt.xlabel('Year')
plt.ylabel('Delivered')
plt.title('Joint Plot')

plt.show()
```
